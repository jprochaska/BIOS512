Use `read_csv` to load the data from the TidyTuesday GitHub repository. The project URL is can be found at [https://github.com/rfordatascience/tidytuesday/blob/master/data/2020/2020-10-27/](https://github.com/rfordatascience/tidytuesday/blob/master/data/2020/2020-10-27/).

```r
library('tidyverse')
df = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/refs/heads/master/data/2020/2020-10-27/wind-turbine.csv')
head(df)
```

Each row in the table represents a single turbine.

Make a bar chart 📊 **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
df_sorted = df |>
    mutate(provinces = province_territory |> fct_infreq())

p = ggplot(df_sorted, aes(y = provinces)) + geom_bar()

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
filtered = df |>
    group_by(manufacturer) |>
    summarize(count = n()) |>
    filter(count >= 10)

new_df = df |>
    mutate(manu_simple = ifelse(manufacturer %in% filtered$manufacturer, manufacturer,'other')) |>
    mutate(manufact_sorted = manu_simple |> fct_infreq() |> fct_relevel("other", after = Inf) |> fct_rev())

g = ggplot(new_df, aes(y = manufact_sorted)) + geom_bar()

g
```

```r
#After writing all of the above confusing code and creating the 
#separate dataset I realized I could simply use fct_lump...
#But I decided to not let the work go to waste and kept it anyway
#While also including this code


newest_df = df |>
    mutate(manufact_sorted = manufacturer |> 
           fct_lump(n = 10) |> 
           fct_infreq() |> 
           fct_relevel("Other", after = Inf) |> 
           fct_rev()
          )

g = ggplot(new_df, aes(y = manufact_sorted)) + geom_bar()

g
```
